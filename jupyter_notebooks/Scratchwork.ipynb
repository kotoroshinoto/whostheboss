```python
from scribe_classifier.data.canada.NOCdb.readers import TitleSet
from scribe_classifier.data.canada.NOCdb.util import TitlePreprocessor
from gensim.models import Phrases, Word2Vec
from gensim.models.phrases import Phraser
import re
from gensim.parsing import preprocessing
```

```python
scribe_title_file = open('source_data/processed/scribe/unique_labels/titles.txt','r', encoding='ascii')
scribe_titles = []
scribe_toks_l = []
unicode_re = re.compile("^u'(.*)'$")
firstline = next(scribe_title_file)
for line in scribe_title_file:
    line = line.rstrip().lstrip()
    mo = unicode_re.match(line)
    if mo:
        mog = mo.groups()
        line = mog[0].encode('utf-8').decode()
        line = u'%s' % line
    line = TitlePreprocessor.preprocess_title_prefixes(line)
    line_slug = TitlePreprocessor.preprocess_slugify(line)
    line_toks = TitlePreprocessor.tokenize_title(line)
    scribe_titles.append(line)
    scribe_toks_l.append(line_toks)
scribe_title_file.close()
```

```python
tset = TitleSet()
tset.add_titles_from_file("source_data/processed/NOC/all_examples")
ttoks = []
for trecord in tset.records:
    ttoks.append(preprocessing.preprocess_string(trecord.title, filters=[
        preprocessing.strip_punctuation,
        preprocessing.remove_stopwords,
        preprocessing.strip_numeric,
        preprocessing.strip_multiple_whitespaces
    ]))
# tset = TitlePreprocessor.preprocess_titleset_split_genders(tset)
# tset = TitlePreprocessor.preprocess_titleset_prefixes(tset)
# tset = TitlePreprocessor.preprocess_slugify_titleset(tset)
# ttoks = TitlePreprocessor.tokenize_titleset(tset)
```

```python
bigram_phrases = Phrases(ttoks)
bigram = Phraser(bigram_phrases)
trigram_phrases = Phrases(bigram[ttoks])
trigram = Phraser(trigram_phrases)
```
